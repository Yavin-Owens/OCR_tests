# Scond Method 

```python
import spacy
import pandas as pd
import re
import numpy as np

# Load the spaCy model and stop words
nlp = spacy.load("en_core_web_sm")
stop_words = spacy.lang.en.stop_words.STOP_WORDS

# Example dataframe
data = {
    "Text": [
        "Hello, my name is John Doe. You can reach me at john.doe@example.com or 020 1234 5678.",
        "Please contact Jane Smith at jane.smith@example.com or +44 7890 123 456 for further information.",
        "For urgent matters, please contact 9876543210.",
        "No entities in this text.",
        np.nan
    ]
}
df = pd.DataFrame(data)

# Regular expressions to extract entities
name_regex = re.compile(r"\b[A-Z][a-z]+\b")
email_regex = re.compile(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b")
phone_regex = re.compile(r"\b\d{10}\b|\b\d{3}\s\d{3}\s\d{4}\b")

# Function to extract entities using regex and spaCy NER
def extract_entities(text):
    if pd.isnull(text):
        return {"PERSON": [], "EMAIL": [], "PHONE": []}
    
    doc = nlp(text)

    # Remove stop words
    filtered_text = ' '.join([token.text for token in doc if not token.is_stop])
    doc_filtered = nlp(filtered_text)
    
    entities = {
        "PERSON": [],
        "EMAIL": [],
        "PHONE": []
    }

    # Extract entities using regex
    entities["PERSON"].extend(re.findall(name_regex, filtered_text))
    entities["EMAIL"].extend(re.findall(email_regex, filtered_text))
    entities["PHONE"].extend(re.findall(phone_regex, filtered_text))

    # Use spaCy NER to categorize entities
    for ent in doc_filtered.ents:
        if ent.label_ == "PERSON" and ent.text not in entities["PERSON"] and (not pd.isnull(ent.text)):
            entities["PERSON"].append(ent.text)
        elif ent.label_ == "EMAIL" and ent.text not in entities["EMAIL"] and (not pd.isnull(ent.text)):
            entities["EMAIL"].append(ent.text)
        elif ent.label_ == "PHONE" and ent.text not in entities["PHONE"] and (not pd.isnull(ent.text)):
            entities["PHONE"].append(ent.text)

    return entities

# Apply entity extraction to the dataframe
df["Entities"] = df["Text"].apply(extract_entities)

# Create separate columns for each entity type
for entity_type in ["PERSON", "EMAIL", "PHONE"]:
    df[entity_type] = df["Entities"].apply(lambda x: x[entity_type])

# Print the extracted entities
for i, row in df.iterrows():
    print(f"Row {i + 1}:")
    print(f"PERSON: {row['PERSON']}")
    print(f"EMAIL: {row['EMAIL']}")
    print(f"PHONE: {row['PHONE']}")
    print()
```

```python
# Create an empty dataframe for pii
pii = pd.DataFrame()

# Iterate over each row in the original dataframe
for i, row in df.iterrows():
    entities = row["Entities"]  # Extracted entities dictionary

    # Iterate over each entity type
    for entity_type in entities:
        values = entities[entity_type]  # List of values for the entity type

        # Append the values to the pii dataframe
        for value in values:
            pii = pii.append({"Entity": entity_type, "Value": value, "OriginalText": row["Text"]}, ignore_index=True)

# Print the pii dataframe
pii
```

# Putting it together

```python
import pandas as pd
import PIL
from PIL import ImageDraw
import easyocr

# Create an instance of EasyOCR
reader = easyocr.Reader(['en'])

# Read Image
im = PIL.Image.open(file_path)

# Doing OCR. Get bounding boxes.
bounds = reader.readtext(file_path, detail=0)  # detail=0 argument will only give text in array

bounds = reader.readtext(file_path, detail=1) 
# Create a dataframe to store the OCR results
ocr_df = pd.DataFrame(bounds, columns=["Text"])

# Visualize bounding boxes on the image
def draw_boxes(image, bounds, color='black', width=1):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

# Draw bounding boxes on the image
im_with_boxes = draw_boxes(im, bounds)

# Display the image with bounding boxes
im_with_boxes.show()

# Print the OCR results
print("Output:")
print(ocr_df)
```
